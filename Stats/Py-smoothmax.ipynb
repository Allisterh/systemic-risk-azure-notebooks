```python
import numpy as np
import math
import matplotlib.pyplot as plt

equity = np.arange(-10, 50, 0.5)

def f(k,x):
    return np.log(1+np.exp(k*x))/k
```

```python
# red dashes, blue squares and green triangles
plt.plot(equity,equity, 'r--', equity, f(1,equity), 'b--', equity, f(5,equity), 'y--',equity,f(50,equity), 'g--',equity,f(.05,equity), 'g--')
plt.show()
```

```python
# red dashes, blue squares and green triangles
plt.plot(equity,1/equity, 'r--', equity, 1/f(.1,equity), 'b--', equity, 1/f(.2,equity), 'y--',equity,1/f(.5,equity), 'g--',equity,1/f(.05,equity), 'g--')
plt.ylim((-2,2))
plt.show()
```
